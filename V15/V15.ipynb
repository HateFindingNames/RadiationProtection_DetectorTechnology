# Gasionisationsdetektoren  
Das Funktionsprinzip von Gasionisationsdetektoren ist die Ausnutzung der ionisierenden Wechselwirkung zwischen der eintretenden, radioaktiven Strahlung und den Teilchen - bzw. genauer den Elektronenschalen - des Füllgases. Befinden sich die Ionen des Füllgases und die freien Elektronen in einem ausreichend starken äußeren elektrischen Feld, lassen sie sich zur Anode bzw. Kathode hin "sortieren" bevor sie dazu in der Lage sind zu rekombinieren, damit Ladungsneutral und letztlich unempfindlich gegenüber dem äußeren E-Feld zu werden.  
Ist das äußere E-Feld nun so stark um die Ladungen derart zu beschleunigen, dass sie ihrerseits durch Kollisionen Gasatomen ionisieren, wird der Gasionisationsdetektor als _Proportionalzähler_ betrieben. Die Bezeichnung rührt von dem Zusammenhang $E \propto N$ wobei $E$ das angelegte Feld und $N$ die Zählrate ist.

# $\alpha$, $\beta$, $\gamma$ - Strahlung  
$\alpha$-Strahlung bezeichnet ionisierende Strahlung, die meist ein Zerfallsprodukt instabiler Kerne ist und seinerseits ein $\frac{4}{2}He$-Kern darstellt.  

$\beta$-Strahlung entsteht ähnlich der $\alpha$-Strahlung bei Kernzerfällen. Hierbei wird jedoch ein Neutron in ein **Proton**, ein **Elektron** und ein **anti-Neutrino** umgewandelt. Auch möglich ist der Zerfall eines Protons unter Emission eines **Neutrons**, eines **Positrons** und eines **Neutrinos**. Es wird somit unterschieden in $\beta^-$-Zerfälle
\begin{equation}
    n \longrightarrow p^+ + e^- + \bar{\nu}
\end{equation}
und $\beta^+$-Zerfälle
\begin{equation}
    p^+ \longrightarrow n + e^+ + \nu
\end{equation}  

$\gamma$-Strahlung zuletzt meint vom Nukleus eines Atoms emmitierte elektromagnetische Strahlung.

# WW von Strahlung mit Materie
(!) Die Wechselwirkung (kurz WW) werden Strahlungen mit Energie von KeV und MeV von Mateire teilweise oder vollständig absorbiert. Die gamma-Strahlungen besitzen als EM-Wellen eine Photonenenergie von etwa 100 keV bis 5 MeV. Die Wahrscheinlichkeit, dass einer der folgende Wechselwirkungsmöglichkeiten auftreten, hängt von der Höhe der Energie.

**Photoeffekt**: EM-Wellen wird von der inneren Elektronenschale übernommen und aus der Atomhülle gestoßen. Bei weiteren Wechselwirkungen kann es Bremsstrahlung auslösen oder andere Atome ionisieren. Durch den Defektelektron in der Atomhülle kann es zu Auger-Elektronen oder Fluoreszenzstrahlung führen.

**Comptoneffekt**: Die EM-Wellen verlieren durch die WW mit ein Elektron an der Außenschale ein Teil deren Energie. Das Elektron wird weggestoßen und die Restenergie der EM-Welle gestreut.

**Paarbildung**: Die WW findet an den Atomkern statt mit eine Photoenergie von ca. 1.022 MeV. Die EM-Welle schlägt aus den Atomkern ein Teilchen-Antiteilchenpaar und kann Bremmsstrahlungen oder Ionisationen erzeugen.

Die alpha- und beta-Strahlungen sind geladene Teilchen. Die WW ist abhängig von der Masse, Ladung und Abstand des einlaufenden  Teilchens zum Wirkungszentrum.

**Anregung**: Die geladene Teilchen geben bei der WW ein Teil der Energie ab und regen die Elektronen der inneren Schalen. Die Anregung des Elektrons zu eine äußere Schale hält kurzzeitig an und emittiert die überschüssige Energie wieder ab.

**Ionisation**: Identisch zu der Anregung, besitzen die geladene Teilchen mehr Energie. Dies führt es zur Abstoßung von Elektronen an der inneren Schalen. Die Leerstelle ersetzt sich durch die Elektronen an der äußeren Schalen und dabei ein Teil deren Energie als EM-Welle.

**Bremsstrahlung**: Entsteht durch die Richtungs- und Geschwindigkeitsänderung der Ladung. Das Teilchen wird durch den Kernfeld unter Energieverlust abgelenkt und ein Teil seiner Energie als EM-Welle aussenden.

**Coulombstreuung**: Wenn die Teilchenenergie so hoch ist, dass das einlaufende Teilchen die Coloumbkräfte der Atomhülle durchsetzen, können sie direkt mit den Atomkernen wechselwirken. Dabei werden sie ohne Energieverlust gestreut.


# (!) Nulleffekt (Nullrate)
(!) Erfolgt durch die Messung von Strahlungen ohne radioaktives Präparat. Dies verhilft es das Identifizieren von Messwerten von einer Quele die angibt, dass die Strahlungen komplett geschwächt wurden.

## Ameritium-241
Das Isotop $^{241}_{95}Am$ ist ein $\alpha$-Strahler nach
\begin{equation}
    ^{241}_{95}Am \longrightarrow ^{237}_{93}Np + \alpha + \gamma 59,54\text{keV}
\end{equation}
mit einer Halbwertszeit von $T_{1/2} = 432,2\text{a}$ und einer Zerfallsenergie von $\approx 5,486 \text{MeV}$. Mit einem atomaren Gewicht von $241,06\text{u}$ enthält $1\text{g}$ $^{241}_{95}Am$ schätzungsweise
\begin{equation}
    1\text{u}^{-1} \cdot {1,661 \cdot 10^{-24}g \cdot 241,06u} \approx 2,5 \cdot 10^{21}
\end{equation}
Kerne.
Somit ist nach $A(t) = \frac{ln(2)}{T_{1/2}} \cdot N_0 \cdot e^{-\frac{ln(2)}{T_{1/2}}t}$ mit etwa $1,27 \cdot 10^{11}$ Zerfällen pro Sekunde zu rechnen.  

Das Zerfallsprodukt $^{237}Np$ besitzt eine Halbwertszeit von $2,144 \cdot 10^6\text{a}$(!) und wird damit mit irrelevant für die Messung angenommen.  

## Strontium-90
Das Isotop $^{90}_{38}Sr$ zerfällt durch $\beta^{-}$-Zerfall mit einer Halbwertszeit von $T_{1/2} = 28,8\text{a}$ und einer Zerfallsenergie von $\approx 0,546 \text{MeV}$. Das Zerfallsprodukt ist Ytterium-90.
\begin{equation}
    ^{90}_{38}St \longrightarrow ^{90}_{39}Y + p^+ + e^- + \bar{\nu}
\end{equation}
Mit einem atomaren Gewicht von $87,62\text{u}$ enthält $1\text{g}$ $^{90}_{38}Sr$
\begin{equation}
    \frac{1u}{1,661 \cdot 10^{-24}g \cdot 87,62u} \approx 6,9 \cdot 10^{21}
\end{equation}
Kerne.
Somit sind nach $A(t) = \frac{ln(2)}{T_{1/2}} \cdot N_0 \cdot e^{-\frac{ln(2)}{T_{1/2}}t}$ mit etwa $5,24 \cdot 10^{12}$ Zerfällen pro Sekunde zu rechnen.

$^{90}_{39}Y$ seinerseits zerfällt mit einer Halbwertszeit von $64.1\text{h}$ unter $\beta^-$-Emission relativ schnell weiter zu Zirconium-90 mit einer Zerfallsenergie von $2,28\text{MeV}$  
Die stöchiometrischen Faktoren ändern sich während der Zerfallsreihe nicht - wohl aber die Zerfallsenergie. Es ist also anzunehmen, dass während der Messung die durch den Zerfall des Spaltproduktes entstehende Strahlung in ihrer Energie dominieren wird. _Mit Blick auf die Zerfallsenergien der jeweiligen Strahler ist jedoch der maximale Messbereich des Kontaminationsmessgerätes zu beachten._

# Aktivität, Dosisleistung
## (!) Aktivität
(!) Die Aktivität einer radioaktiven Probe gibt den Erwartungswert der umgewandelten Teilchen über eine Zeit. Die Einheit der Aktivität lautet Becquerel $Bq$ und gibt die Anzahl der Umwandlungen pro Sekunde. Ein weiterer Part über die Aktivität ist die Zerfallskonstante $\lambda$. Die Zerfallskonstante ist materialspezifisch.
\begin{equation}
    A(t) = A_{0} \cdot e^{- \lambda \cdot t}
\end{equation}
## (!) Dosisleistung
(!) Beruht auf die Absorption von Strahlungsenergie mit Materie. Die Menge wird in Joule durch Kilogramm $J / kg$, auch als Gray $Gy$ genannt, angegeben. Die Absorption der Energie ist bei gleicher Strahlungsintensität  in jede Materie unterschiedlich. Deshalb muss bei der Energiedosisangabe immer das Absorbermaterial dazu gennant werden.

# Absorptionsgesetz (Gültigkeit für $\alpha$ , $\beta$ , $\gamma$ -  Strahlung
(!) Die Zählrate von einen Quellaparat ändert über die Dicke des Materials. Hinzu hängt es von der Schwächungskoeffizient $\mu$. Es setzt sich additiv aus den Koeffizienten für einzelnen Wechselwirkung  zusammen. 
\begin{equation}
    I(d) = I_{0} \cdot e^{- \mu \cdot d}
\end{equation}
Die einzelne Komponente sind Energieabhängig und zeigt eine komplexe lineare Schwächungsverlauf mit der Photonenenergie. Er ist eine für das jeweilige Absorbermaterial wohl definierte und charakterische Größe.

```python
# Pakete aufrufen
import matplotlib.pyplot as plt
from matplotlib import pyplot
import pandas as pd
from pandas import read_csv
from numpy import arange
import numpy as np
from scipy.optimize import curve_fit

# Definitionen zum Fitten
def objective(x, a, b):
	return a * np.exp(-x * b)

# den "Weg" und das Öffnen der Dateien

paths = ["./Messungen/Am_A1.csv",
        "./Messungen/Am_A2.csv",
        "./Messungen/Co_A2.csv",
        "./Messungen/Sr_A1.csv",
        "./Messungen/Sr_A2.csv",]

with open(paths[0], newline='' , encoding='utf-8-sig'):
    am1 = read_csv(paths[0], delimiter=';')
with open(paths[1], newline='' , encoding='utf-8-sig'):
    am2 = read_csv(paths[1], delimiter=';')
with open(paths[2], newline='' , encoding='utf-8-sig'):
    co2 = read_csv(paths[2], delimiter=';')
with open(paths[3], newline='' , encoding='utf-8-sig'):
    sr1 = read_csv(paths[3], delimiter=';') 
with open(paths[4], newline='' , encoding='utf-8-sig'):
    sr2 = read_csv(paths[4], delimiter=';')   
```
